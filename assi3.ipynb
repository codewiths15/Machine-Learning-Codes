```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

# Gradient Descent parameters
x = 2            # Starting point
learning_rate = 0.1  # Step size
num_iterations = 50  # Number of iterations

# Lists to store values for plotting
x_values = [x]
y_values = [function(x)]

# Perform Gradient Descent
for i in range(num_iterations):
    grad = derivative(x)          # Calculate the gradient
    x = x - learning_rate * grad   # Update x to move towards minimum
    x_values.append(x)             # Store x for plotting
    y_values.append(function(x))   # Store y for plotting
    print(f"Step {i+1}: x = {x}, y = {function(x)}")

print(f"\nLocal minimum at x = {x}, y = {function(x)}")

# Plotting the function and descent path
x_range = np.linspace(-5, 5, 100)  # Range of x values for smooth curve
y_range = function(x_range)        # Corresponding y values

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="y = (x + 3)^2", color="blue")
plt.scatter(x_values, y_values, color="red", label="Descent Path")
plt.plot(x_values, y_values, color="red", linestyle="--", marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minimum of y = (x + 3)^2")
plt.legend()
plt.show()
```
