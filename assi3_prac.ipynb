```python
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    return (x+3)**2

def dervative(x):
    return 2*(x+3)

x=2
iteration_limit=50
learning_rate=0.1

x_values=[x]
y_values=[function(x)]

for i in range(iteration_limit):
    grad= dervative(x)
    x=x-learning_rate*grad
    x_values.append(x)
    y_values.append(function(x))
    print(f"The value of x is {x} and y is {function(x)}")
print(f"The minimum value of x is {x} and y is {function(x)}")

x_range = np.linspace(-5,5,100)
y_range=function(x_range)

plt.figure(figsize=(10,6))
plt.plot(x_range,y_range,color="blue", label="(x+3)**2")
plt.scatter(x_values,y_values,color="red", label="Descent path")
plt.plot(x_values,y_values,color="red", linestyle="--", marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent algorithm")
plt.legend()
plt.show()
```
