```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding="ISO-8859-1")
df.head()
```

```python
features=df[['QUANTITYORDERED','SALES']]
features=features.dropna()
```

```python
print(features)
```

```python
wcss=[]

for k in range(1,11):
    k_means=KMeans(n_clusters=k, init="k-means++", random_state=42)
    k_means.fit(features)
    wcss.append(k_means.inertia_)
```

```python
print(wcss)
```

```python
plt.plot(range(1,11), wcss)
plt.title("Elbow method")
plt.xlabel("wcss value")
plt.ylabel("Number of clusters")
plt.show()
```

```python
k_means=KMeans(n_clusters=3, init="k-means++", random_state=42)
features["Clusters"]=k_means.fit_predict(features)
```

```python
plt.scatter(features['QUANTITYORDERED'], features['SALES'], c=features['Clusters'], cmap="viridis")
plt.title("Clusters plotting")
plt.xlabel("Quantity ordered")
plt.ylabel("Sales")
plt.show()
```
